```python
# ê¸°ë³¸
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl

# ê·¸ë˜í”„ ì„¤ì •
sns.set()

# í°íŠ¸ ì„¤ì •
mpl.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14

# [2] ì‚¬ìš©í•  ì»¬ëŸ¼ê³¼ ë°ì´í„° ì¤€ë¹„
cols = ['ì¦ê°ìœ¨_ì´ìš©ê±´ìˆ˜_ì‹ ìš©_ì „ì›”', 'ì¦ê°ìœ¨_ì´ìš©ê¸ˆì•¡_ì‹ ìš©_ì „ì›”']
segment_means = df.groupby('Segment')[cols].mean().reset_index()

# (1) ì‹ ìš© ì´ìš©ê±´ìˆ˜ ì¦ê°ìœ¨
plt.subplot(1, 2, 1)
sns.barplot(
    x='Segment',
    y='ì¦ê°ìœ¨_ì´ìš©ê±´ìˆ˜_ì‹ ìš©_ì „ì›”',
    hue='Segment',
    data=segment_means,
    palette='Blues_d',
    legend=False
)
plt.title('ì„¸ê·¸ë¨¼íŠ¸ë³„ ì‹ ìš© ì´ìš©ê±´ìˆ˜ ì¦ê°ìœ¨ (ì „ì›” ëŒ€ë¹„)')
plt.ylabel('ì¦ê°ìœ¨')
plt.ylim(segment_means['ì¦ê°ìœ¨_ì´ìš©ê±´ìˆ˜_ì‹ ìš©_ì „ì›”'].min() * 1.2, 0)  # ìŒìˆ˜ ì¤‘ì‹¬

# (2) ì‹ ìš© ì´ìš©ê¸ˆì•¡ ì¦ê°ìœ¨
plt.subplot(1, 2, 2)
sns.barplot(
    x='Segment',
    y='ì¦ê°ìœ¨_ì´ìš©ê¸ˆì•¡_ì‹ ìš©_ì „ì›”',
    hue='Segment',
    data=segment_means,
    palette='Greens_d',
    legend=False
)
plt.title('ì„¸ê·¸ë¨¼íŠ¸ë³„ ì‹ ìš© ì´ìš©ê¸ˆì•¡ ì¦ê°ìœ¨ (ì „ì›” ëŒ€ë¹„)')
plt.ylabel('ì¦ê°ìœ¨')
plt.ylim(0, segment_means['ì¦ê°ìœ¨_ì´ìš©ê¸ˆì•¡_ì‹ ìš©_ì „ì›”'].max() * 1.2)  # ì–‘ìˆ˜ ì¤‘ì‹¬

plt.tight_layout()
plt.show()
```
