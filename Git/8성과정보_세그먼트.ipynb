```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib as mpl

# 그래프 설정
sns.set()

# 폰트 설정
mpl.rcParams['font.family'] = 'NanumGothic'
mpl.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14

# [2] 사용할 컬럼과 데이터 준비
cols = ['증감율_이용건수_신용_전월', '증감율_이용금액_신용_전월']
segment_means = df.groupby('Segment')[cols].mean().reset_index()

# (1) 신용 이용건수 증감율
plt.subplot(1, 2, 1)
sns.barplot(
    x='Segment',
    y='증감율_이용건수_신용_전월',
    hue='Segment',
    data=segment_means,
    palette='Blues_d',
    legend=False
)
plt.title('세그먼트별 신용 이용건수 증감율 (전월 대비)')
plt.ylabel('증감율')
plt.ylim(segment_means['증감율_이용건수_신용_전월'].min() * 1.2, 0)  # 음수 중심

# (2) 신용 이용금액 증감율
plt.subplot(1, 2, 2)
sns.barplot(
    x='Segment',
    y='증감율_이용금액_신용_전월',
    hue='Segment',
    data=segment_means,
    palette='Greens_d',
    legend=False
)
plt.title('세그먼트별 신용 이용금액 증감율 (전월 대비)')
plt.ylabel('증감율')
plt.ylim(0, segment_means['증감율_이용금액_신용_전월'].max() * 1.2)  # 양수 중심

plt.tight_layout()
plt.show()
```
