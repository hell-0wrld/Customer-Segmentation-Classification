```python
# 전체 컬럼

summary = []

for col in merged_approval.columns:  # 컬럼명 순서 그대로
    # 결측치 비율
    null_pct = null_summary.loc[col, "결측치비율"] * 100 if col in null_summary.index else 0

    # 0 비율
    zero_pct = zero_ratio.loc[col] * 100 if col in zero_ratio.index else 0

    # 데이터 타입
    dtype = merged_approval[col].dtype

    # 샘플값 (NaN 제외)
    samples = merged_approval[col].dropna().unique()[:5]

    summary.append({
        "컬럼명": col,
        "결측치비율 (%)": round(null_pct, 2),
        "0 비율 (%)": round(zero_pct, 2),
        "TYPE": str(dtype),
        "샘플값": samples
    })

# DataFrame 생성
import pandas as pd
result_df = pd.DataFrame(summary)

# 번호 컬럼 추가 (1부터 시작)
result_df.insert(0, "번호", range(1, len(result_df) + 1))

from IPython.display import display
display(result_df)
```

```python
# 표준 편차

# 숫자형 컬럼 선택
numeric_cols = merged_approval.select_dtypes(include="number").columns

# 결과 저장 리스트
summary = []

# 기준 (표준편차 몇 배를 이상치로 볼지)
k = 2

for col in numeric_cols:
    series = merged_approval[col].dropna()
    if series.empty:
        continue

    var = series.var()
    std = series.std()
    mean = series.mean()

    lower = mean - k * std
    upper = mean + k * std

    outliers = ((series < lower) | (series > upper)).sum()
    outlier_ratio = outliers / len(series) * 100

    summary.append({
        "컬럼명": col,
        "분산": round(var, 2),
        "표준편차": round(std, 2),
        "이상치 개수": outliers,
        "이상치 비율 (%)": round(outlier_ratio, 2)
    })

# DataFrame으로 보기
result_df = pd.DataFrame(summary).sort_values("이상치 비율 (%)", ascending=False)

from IPython.display import display
display(result_df)
```

```python
# 분산과 표준편차가 모두 0인 컬럼

epsilon = 1e-6  # 허용 오차 (아주 작은 값)

zero_variance_cols = result_df[
    (result_df["분산"].abs() <= epsilon) &
    (result_df["표준편차"].abs() <= epsilon)
]["컬럼명"].tolist()

print("분산과 표준편차가 사실상 0인 컬럼:")
for col in zero_variance_cols:
    print("-", col)

print(f"\n총 {len(zero_variance_cols)}개 컬럼")
```
