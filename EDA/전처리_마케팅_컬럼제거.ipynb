```python
def natural_key(text):
    return [int(t) if t.isdigit() else t.lower() for t in re.split(r'(\d+)', str(text))]

# ì»¬ëŸ¼ ëª©ë¡ (object, int, float í¬í•¨)
object_cols = merged_df.select_dtypes(include=['float', 'object']).columns.tolist()

for col in object_cols:
    # ê³ ìœ ê°’ ì„¸ê¸° (dropna=Falseë¡œ None í¬í•¨)
    value_counts = merged_df[col].value_counts(dropna=False)

    # í‘œì‹œìš© ì¸ë±ìŠ¤ ê°€ê³µ
    def convert_val(val):
        if pd.isna(val):
            return 'NaN'
        elif val is None:
            return 'None'
        else:
            return str(val)

    value_counts_display = value_counts.copy()
    value_counts_display.index = value_counts_display.index.map(convert_val)

    # ì •ë ¬
    sorted_vals = sorted(value_counts_display.index.tolist(), key=natural_key)
    unique_count = len(value_counts_display)

    print(f"ì»¬ëŸ¼ëª…: {col} (ê³ ìœ ê°’ {unique_count}ê°œ)")

    # ì¶œë ¥ ê°œìˆ˜ ì¡°ì •
    if unique_count > 10:
        display_vals = sorted_vals[:5] + ["..."] + sorted_vals[-5:]
    else:
        display_vals = sorted_vals

    for val in display_vals:
        if val == "...":
            print(" - ...")
        else:
            count = value_counts_display[val]
            print(f"  {val}: {count}ê±´")

    # ê²°ì¸¡ì¹˜(NaN + None) ê°œìˆ˜ ê³„ì‚°
    nan_count = merged_df[col].apply(lambda x: pd.isna(x)).sum()
    none_count = merged_df[col].apply(lambda x: x is None).sum()
    total_missing = nan_count + none_count

    print(f" - ì „ì²´ ê²°ì¸¡ì¹˜ : {total_missing}ê±´")
    print("-" * 60)
```

```python
def check_column_name_issues(df):
    issues = {
        'ë„ì–´ì“°ê¸° í¬í•¨': [],
        'íŠ¹ìˆ˜ë¬¸ì í¬í•¨': [],
        'ìˆ«ìë¡œ ì‹œì‘': [],
        'ì»¬ëŸ¼ëª… ì¤‘ë³µ': []
    }

    import re
    from collections import Counter

    # ì¤‘ë³µ ì»¬ëŸ¼ í™•ì¸
    col_counts = Counter(df.columns)
    issues['ì»¬ëŸ¼ëª… ì¤‘ë³µ'] = [col for col, count in col_counts.items() if count > 1]

    for col in df.columns:
        if ' ' in col:
            issues['ë„ì–´ì“°ê¸° í¬í•¨'].append(col)
        if re.search(r'[^\w]', col):  # ì˜ìˆ«ì + _ ì œì™¸í•œ íŠ¹ìˆ˜ë¬¸ì í¬í•¨
            issues['íŠ¹ìˆ˜ë¬¸ì í¬í•¨'].append(col)
        if re.match(r'^\d', col):
            issues['ìˆ«ìë¡œ ì‹œì‘'].append(col)

    # ê²°ê³¼ ì¶œë ¥
    print("ğŸ” ì»¬ëŸ¼ëª… ì´ìƒ íŒ¨í„´ íƒì§€ ê²°ê³¼:")
    for key, cols in issues.items():
        if cols:
            print(f"\nâš ï¸ {key} ì»¬ëŸ¼ ({len(cols)}ê°œ):")
            for c in cols:
                print(f" - {c}")
        else:
            print(f"\nâœ… {key} ì—†ìŒ")

# ì‚¬ìš© ì˜ˆì‹œ
check_column_name_issues(merged_df)
```
