```python
# 파일 불러오기
merged_df = pd.read_parquet('/content/drive/MyDrive/11_파이널프로젝트관련/file/채널정보_세그먼트병합.parquet')

# 컬럼이 생략되지 않도록 출력 옵션 설정
pd.set_option('display.max_columns', None)  # 컬럼 모두 출력
pd.set_option('display.width', None)        # 가로 너비 제한 해제
pd.set_option('display.max_colwidth', None) # 컬럼 내용 길이 제한 해제

# 데이터 상위 5개 행 출력
merged_df.head()
```

## 인코딩

```python
import re

def natural_key(text):
    # 문자열을 숫자/문자 부분으로 분리해서 리스트로 반환
    return [int(t) if t.isdigit() else t.lower() for t in re.split('(\d+)', text)]

object_cols = merged_df.select_dtypes(include='object').columns.tolist()

for col in object_cols:
    unique_vals = merged_df[col].fillna('None').unique()
    unique_vals_sorted = sorted(unique_vals, key=natural_key)
    unique_count = len(unique_vals_sorted)

    if unique_count > 10:
        display_vals = unique_vals_sorted[:5] + ["..."] + unique_vals_sorted[-5:]
    else:
        display_vals = unique_vals_sorted

    print(f"📝 컬럼명: {col} (고유값 {unique_count}개)")
    print(f" - 고유값 : {display_vals}")
    print("-" * 60)
```

```python
def natural_key(text):
    return [int(t) if t.isdigit() else t.lower() for t in re.split(r'(\d+)', str(text))]

# 컬럼 목록 (object, int, float 포함)
object_cols = merged_df.select_dtypes(include=['float', 'object']).columns.tolist()

for col in object_cols:
    # 고유값 세기 (dropna=False로 None 포함)
    value_counts = merged_df[col].value_counts(dropna=False)

    # 표시용 인덱스 가공
    def convert_val(val):
        if pd.isna(val):
            return 'NaN'
        elif val is None:
            return 'None'
        else:
            return str(val)

    value_counts_display = value_counts.copy()
    value_counts_display.index = value_counts_display.index.map(convert_val)

    # 정렬
    sorted_vals = sorted(value_counts_display.index.tolist(), key=natural_key)
    unique_count = len(value_counts_display)

    print(f"컬럼명: {col} (고유값 {unique_count}개)")

    # 출력 개수 조정
    if unique_count > 10:
        display_vals = sorted_vals[:5] + ["..."] + sorted_vals[-5:]
    else:
        display_vals = sorted_vals

    for val in display_vals:
        if val == "...":
            print(" - ...")
        else:
            count = value_counts_display[val]
            print(f"  {val}: {count}건")

    # 결측치(NaN + None) 개수 계산
    nan_count = merged_df[col].apply(lambda x: pd.isna(x)).sum()
    none_count = merged_df[col].apply(lambda x: x is None).sum()
    total_missing = nan_count + none_count

    print(f" - 전체 결측치 : {total_missing}건")
    print("-" * 60)
```

## 오류 확인

```python
def check_column_name_issues(df):
    issues = {
        '띄어쓰기 포함': [],
        '특수문자 포함': [],
        '숫자로 시작': [],
        '컬럼명 중복': []
    }

    import re
    from collections import Counter

    # 중복 컬럼 확인
    col_counts = Counter(df.columns)
    issues['컬럼명 중복'] = [col for col, count in col_counts.items() if count > 1]

    for col in df.columns:
        if ' ' in col:
            issues['띄어쓰기 포함'].append(col)
        if re.search(r'[^\w]', col):  # 영숫자 + _ 제외한 특수문자 포함
            issues['특수문자 포함'].append(col)
        if re.match(r'^\d', col):
            issues['숫자로 시작'].append(col)

    # 결과 출력
    print("🔍 컬럼명 이상 패턴 탐지 결과:")
    for key, cols in issues.items():
        if cols:
            print(f"\n⚠️ {key} 컬럼 ({len(cols)}개):")
            for c in cols:
                print(f" - {c}")
        else:
            print(f"\n✅ {key} 없음")

# 사용 예시
check_column_name_issues(merged_df)
```
