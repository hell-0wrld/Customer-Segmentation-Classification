```python
# 이상치 컬럼 확인

# 숫자형 컬럼만
numeric_cols = merged_approval.select_dtypes(include='number').columns

# 결과 저장
outlier_summary = []

for col in numeric_cols:
    series = merged_approval[col].dropna()
    if series.empty:
        continue

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((series < lower_bound) | (series > upper_bound)).sum()
    outlier_ratio = (outlier_count / len(series)) * 100

    outlier_summary.append({
        "컬럼명": col,
        "하한": round(lower_bound, 2),
        "상한": round(upper_bound, 2),
        "이상치 개수": outlier_count,
        "이상치 비율 (%)": round(outlier_ratio, 2)
    })

# DataFrame으로 보기
outlier_df = pd.DataFrame(outlier_summary).sort_values("이상치 비율 (%)", ascending=False)

from IPython.display import display
display(outlier_df)
```

```python
# 숫자형 컬럼만 선택
numeric_cols = merged_approval.select_dtypes(include="number").columns

# 분산 계산
variance_series = merged_approval[numeric_cols].var()

# 표로 변환
variance_df = pd.DataFrame({
    "분산": variance_series
}).sort_values("분산", ascending=False)

from IPython.display import display
display(variance_df)
```

```python
# 표준 편차

# 숫자형 컬럼 선택
numeric_cols = merged_approval.select_dtypes(include="number").columns

# 결과 저장 리스트
summary = []

# 기준 (표준편차 몇 배를 이상치로 볼지)
k = 2

for col in numeric_cols:
    series = merged_approval[col].dropna()
    if series.empty:
        continue

    var = series.var()
    std = series.std()
    mean = series.mean()

    lower = mean - k * std
    upper = mean + k * std

    outliers = ((series < lower) | (series > upper)).sum()
    outlier_ratio = outliers / len(series) * 100

    summary.append({
        "컬럼명": col,
        "분산": round(var, 2),
        "표준편차": round(std, 2),
        "이상치 개수": outliers,
        "이상치 비율 (%)": round(outlier_ratio, 2)
    })

# DataFrame으로 보기
result_df = pd.DataFrame(summary).sort_values("이상치 비율 (%)", ascending=False)

from IPython.display import display
display(result_df)
```

```python
# 분산과 표준편차가 모두 0인 컬럼

zero_variance_cols = result_df[
    (result_df["분산"] == 0) & (result_df["표준편차"] == 0)
]["컬럼명"].tolist()

# 결과 출력
print("분산과 표준편차가 모두 0인 컬럼:")
for col in zero_variance_cols:
    print("-", col)

print(f"\n총 {len(zero_variance_cols)}개 컬럼")
```
