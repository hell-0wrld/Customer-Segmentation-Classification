```python
# 고유값 개수 확인
single_value_cols = []
print("\n--- 고정된 값 확인 ---")
for col in df.columns:
    unique_value_count = df[col].nunique(dropna=False)

    # 고유한 값의 개수가 1개인 경우
    if unique_value_count == 1:
        single_value_cols.append(col)

print(f"\n고정값 : \n{single_value_cols}")


# 컬럼 값이 완전히 동일한 중복 컬럼 찾기
# 병합 후 재실행
print("\n--- 중복 컬럼 확인 ---")
duplicated_cols = {}
cols = df.columns.tolist()
for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        col1 = cols[i]
        col2 = cols[j]
        if df[col1].equals(df[col2]):
            if col1 not in duplicated_cols:
                duplicated_cols[col1] = []
            duplicated_cols[col1].append(col2)
if duplicated_cols:
    print(f"중복 컬럼들: \n{duplicated_cols}")
```

# 2. 🖥️ 디지털 채널 이용 성향

```python
# 1. '1회 이상', '10회 이상' 등 텍스트를 숫자로 변환하는 함수
def convert_visit_counts(val):
    if isinstance(val, str):
        if val == '10회 이상':
            return 10
        elif val == '1회 이상':
            return 1
        elif val.isdigit():
            return int(val)
        else:
            return 0
    elif pd.isna(val):
        return 0
    else:
        return val

# 2. 사용할 컬럼 리스트
cols_2 = ['방문횟수_앱_R6M', '방문횟수_PC_R6M', '방문횟수_모바일웹_R6M']

# 3. 각 컬럼에 변환 함수 적용
for col in cols_2:
    if col in df.columns:
        df[col] = df[col].apply(convert_visit_counts)

# 4. 세그먼트별 평균 계산
segment_means = df.groupby('Segment')[cols_2].mean()
print("✅ 세그먼트별 방문횟수 평균:\n", segment_means)
```

```python
# 1. 앞서 사용한 방문횟수 변환 함수 (필요하면 다시 적용)
def convert_visit_counts(val):
    if isinstance(val, str):
        if val == '10회 이상':
            return 10
        elif val == '1회 이상':
            return 1
        elif val.isdigit():
            return int(val)
        else:
            return 0
    elif pd.isna(val):
        return 0
    else:
        return val

cols = ['방문횟수_앱_R6M', '방문횟수_PC_R6M', '방문횟수_모바일웹_R6M']

for col in cols:
    if col in df.columns:
        df[col] = df[col].apply(convert_visit_counts)

# 2. 디지털 채널 방문횟수 합계 컬럼 생성
df['디지털채널_방문횟수_합계'] = df[cols].sum(axis=1)

# 3. 세그먼트별 평균 계산
segment_means_sum = df.groupby('Segment')['디지털채널_방문횟수_합계'].mean()
print("✅ 세그먼트별 디지털 채널 방문횟수 합계 평균:\n", segment_means_sum)
```

# 5. 📲 OS 구분

```python
# None (또는 NaN)인 행만 필터링
cols_5 = df[df['OS구분코드'].isnull()]

# 세그먼트별로 개수 집계
result = cols_5.groupby('Segment').size().reset_index(name='None_개수')

print(result)
```

```python
# 전체 기준: 세그먼트별 전체 건수
total_counts = df.groupby('Segment').size().reset_index(name='Segment수')

# None 기준: OS구분코드가 None인 행만 필터링
cols_5 = df[df['OS구분코드'].isnull()]
none_counts = cols_5.groupby('Segment').size().reset_index(name='None_건수')

# 병합하여 비율 계산
merged = pd.merge(total_counts, none_counts, on='Segment', how='left')
merged['None_건수'] = merged['None_건수'].fillna(0)  # None 없는 세그먼트는 0으로
merged['None_비율(%)'] = ((merged['None_건수'] / merged['Segment수']) * 100).round(2)

print(merged)
```
